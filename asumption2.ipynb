```python
id_list = []
conn = db.connect()
post_list = db.executeSQL(conn,"select id,published_at from yahoo_posts where title like '%裴洛西%' or summary like '%裴洛西%' order by comment_num DESC",fetch=True)
```

```python
window = 60 * 60 #window set to 1 hour from post 
time_diff_list = []
for id,time in post_list:
    endtime = time + window
    time_diff = db.executeSQL(conn,f"select createdat from yahoo_comments where contextid = '{id}' and createdat < {endtime} order by createdat ASC",fetch=True)
    time_diff = [x-time for (x,) in time_diff]
    time_diff_list.append(time_diff)


    
```
