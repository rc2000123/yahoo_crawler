```python
import requests
from collections import Counter

import pandas as pd
```

```python
news_data = []

for offset in range(0, 5000, 10):
    print(offset)
    url =f'''
    https://tw.news.yahoo.com/_td-news/api/resource/NewsSearchService;loadMore=true;mrs=%7B%22size%22%3A%7B%22w%22%3A220%2C%22h%22%3A128%7D%7D;offset={offset}\
    ;query=%E7%83%8F%E5%85%8B%E8%98%AD;useNCP=false;usePrefetch=false?bkt=news-TW-zh-Hant-TW-def&device=desktop&ecma=modern&feature=enableOlympicsModule%2CoathPlayer%2Carticle20%2CvideoDocking&intl=tw&lang=zh-Hant-TW&partner=none&prid=6c9atqph1le1g&region=TW&site=news&tz=Asia%2FTaipei&ver=2.3.1912&returnMeta=true
    '''.replace('    ', '')
    rs = requests.get(url)
    news_data.extend(rs.json()['data'])
```
