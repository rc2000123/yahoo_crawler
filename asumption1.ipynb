```python
 
import dbconnect as db
```

```python
id_list = []
conn = db.connect()
res = db.executeSQL(conn,"select id from yahoo_posts where title like '%裴洛西%' or summary like '%裴洛西%' order by comment_num DESC",fetch=True)
id_list = [x for (x,) in res]
```

```python
user_dict = {}
for id in id_list:
    res = db.executeSQL(conn,f"select author_guid from yahoo_comments where contextid='{id}'",fetch=True)
    user_list = [x for (x,) in res]
    user_list = list(set(user_list))
    for user in user_list:
        if user not in user_dict:
            user_dict[user] = 1
        else:
            user_dict[user] += 1 
    
```

```python
user_dict
```

```python
user_list=list(user_dict.values())

avg = sum(user_list) / len(user_list)
```
